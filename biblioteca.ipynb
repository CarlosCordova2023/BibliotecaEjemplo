```python
def agregar_libro(biblioteca, titulo, autor, año):
    if titulo not in biblioteca:
        biblioteca[titulo] = (autor, año)
        print(f"El libro '{titulo}' ha sido agregado a la biblioteca.")
    else:
        print(f"El libro '{titulo}' ya existe en la biblioteca.")

def buscar_libro(biblioteca, titulo):
    if titulo in biblioteca:
        autor, año = biblioteca[titulo]
        print(f"Detalles del libro '{titulo}': Autor: {autor}, Año: {año}")
    else:
        print(f"El libro '{titulo}' no está en la biblioteca.")

def eliminar_libro(biblioteca, titulo):
    if titulo in biblioteca:
        del biblioteca[titulo]
        print(f"El libro '{titulo}' ha sido eliminado de la biblioteca.")
    else:
        print(f"El libro '{titulo}' no está en la biblioteca.")

def listar_libros(biblioteca):
    if biblioteca:
        print("Libros en la biblioteca:")
        for titulo, (autor, año) in biblioteca.items():
            print(f"Título: {titulo}, Autor: {autor}, Año: {año}")
    else:
        print("La biblioteca está vacía.")

def listar_autores_unicos(biblioteca):
    autores = {autor for autor, año in biblioteca.values()}
    if autores:
        print("Autores únicos en la biblioteca:")
        for autor in autores:
            print(autor)
    else:
        print("No hay autores en la biblioteca.")

# Ejemplo de uso
biblioteca = {}

while True:
    print("\nOpciones:")
    print("1. Agregar un libro")
    print("2. Buscar un libro")
    print("3. Eliminar un libro")
    print("4. Listar todos los libros")
    print("5. Listar autores únicos")
    print("6. Salir")

    opcion = input("Seleccione una opción (1-6): ")

    if opcion == "1":
        titulo = input("Ingrese el título del libro: ")
        autor = input("Ingrese el autor del libro: ")
        año = input("Ingrese el año de publicación del libro: ")
        agregar_libro(biblioteca, titulo, autor, año)

    elif opcion == "2":
        titulo = input("Ingrese el título del libro que desea buscar: ")
        buscar_libro(biblioteca, titulo)

    elif opcion == "3":
        titulo = input("Ingrese el título del libro que desea eliminar: ")
        eliminar_libro(biblioteca, titulo)

    elif opcion == "4":
        listar_libros(biblioteca)

    elif opcion == "5":
        listar_autores_unicos(biblioteca)

    elif opcion == "6":
        print("Saliendo del programa.")
        break

    else:
        print("Opción no válida. Intente de nuevo.")
```
